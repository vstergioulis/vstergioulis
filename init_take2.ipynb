```python
import os
import pandas as pd
import sklearn
from sklearn.preprocessing import OneHotEncoder
```

```python
def dataset_dataframe():
    paths = []
    labels_em = []
    labels_actor=[]
    labels_phrase=[]
    labels_version=[]

    for dirname , _ , filenames in os.walk('Dataset/wav'):
        for filename in filenames:
            paths.append(os.path.join(dirname , filename))
            
            label_em = filename[5:6]
            label_em = label_em.split('.')[0]
            labels_em.append(label_em)

            label_actor=filename[:2]
            label_actor = label_actor.split('.')[0]
            labels_actor.append(label_actor)

            label_phrase=filename[2:5]
            label_phrase = label_phrase.split('.')[0]
            labels_phrase.append(label_phrase)

            label_version=filename[6:]
            label_version = label_version.split('.')[0]
            labels_version.append(label_version)

    #Creating the Dataframe
    
    df = pd.DataFrame()
    df['SOUND'] = paths
    df['LABEL'] = labels_em
    df['ACTOR'] = labels_actor
    df['PHRASE'] = labels_phrase
    df['VERSION'] = labels_version

    # Transliterate from german to english
    #1 Emotions
    df['LABEL'] = df['LABEL'].replace({'F': 'Happiness', 'W': 'Anger', 'L': 'Boredomm', 'E': 'Disgust' , 'A': 'Fear' ,'T': 'Sadness' ,'N': 'Neutral' })

    #2 Phrases
    df['PHRASE'] = df['PHRASE'].replace({'a01': 'The tablecloth is lying on the frigde.', 'a02': 'She will hand it in on Wednesday.', 'a04': 'Tonight I could tell him.', 'a05': 'The black sheet of paper is located up there besides the piece of timber.' , 'a07': 'In seven hours it will be.' ,'b01': 'What about the bags standing there under the table?' ,'b02': 'They just carried it upstairs and now they are going down again.' , 'b03' : 'Currently at the weekends I always went home and saw Agnes.' , 'b09' : 'I will just discard this and then go for a drink with Karl.'  , 'b10' : 'It will be in the place where we always store it.'})
    df['ACTOR'] = df['ACTOR'].replace({'03' : 'male,31' , '08' : 'female,34' , '09' : 'female,21' , '10' : 'male,32' , '11' : 'male,26' , '12': 'male,30' , '13' : 'female,32' , '14' : 'female,35' , '15' : 'male,25' , '16' : 'female,31'})
    df[['GENDER', 'AGE']] = df['ACTOR'].str.split(',', expand=True)

    df = df.drop(columns = 'ACTOR')

    # Perform one hot encoding for the gender stuff
    sex_d = pd.get_dummies(df['GENDER'])
    df = pd.concat([df,sex_d], axis=1)
    df = df.drop(['GENDER'], axis=1)

    # Here  I shall point out that everything , even the numbers are strings . THe only exception the boolean gender values
    
    
    print('Now I have Become death , nahh jk dataset is loaded (theoretically))')

    return(df)
```

```python
df = dataset_dataframe()
```
